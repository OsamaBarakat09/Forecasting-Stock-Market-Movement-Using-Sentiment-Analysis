```python
def is_url(input_string):
    try:
        url_pattern = re.compile(
            r'^(https?://)?'                      # Optional scheme (http or https)
            r'((([a-zA-Z0-9-]+\.)+[a-zA-Z]{2,})|'  # Domain name
            r'localhost|'                          # OR localhost
            r'(\d{1,3}\.){3}\d{1,3})'              # OR IPv4 address
            r'(:\d+)?'                             # Optional port
            r'(/[-a-zA-Z0-9@:%._+~#=]*)*'          # Optional path
            r'(\?[;&a-zA-Z0-9%._+~#=-]*)?'         # Optional query
            r'(#[-a-zA-Z0-9_]*)?$'                 # Optional fragment
        )
        return bool(url_pattern.match(input_string))
    except Exception as e:
        pass
```

```python
def extract_date_as_date_type(text):
    # Regex pattern for Month DD, YYYY
    date_pattern = r'\b(\w+)\s(\d{1,2}),\s(\d{4})\b'
    match = re.search(date_pattern, text)
    
    if match:
        # Extract month, day, year
        month_name, day, year = match.groups()
        # Convert to datetime.date
        date_obj = datetime.strptime(f"{month_name} {day} {year}", "%B %d %Y").date()
        return date_obj
    return 0
```

```python
def is_article(url):
    try:
        # Fetch the URL content
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        # Parse the content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Look for significant content in <p> tags (paragraphs)
        paragraphs = soup.find_all('p')
        word_count = sum(len(p.get_text().split()) for p in paragraphs)

        # Check for meta tags indicating an article
        meta_tags = soup.find_all('meta', attrs={'name': ['description', 'keywords']})
        is_article_meta = any('article' in (meta.get('content') or '').lower() for meta in meta_tags)

        # Decide based on word count and meta tags
        if word_count > 200 or is_article_meta:
            return True
        else:
            return False
    except requests.RequestException as e:
        print(f"Error fetching URL: {e}")
        return False
```

```python
import requests
from bs4 import BeautifulSoup

def get_all_classes(url):
    """
    Fetches all unique class names from a webpage.

    Args:
        url (str): The URL of the webpage to scrape.

    Returns:
        set: A set of unique class names found on the page.
    """
    try:
        # Fetch the webpage content
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        response.raise_for_status()  # Raise an exception for HTTP errors
        
        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all elements with a "class" attribute
        elements_with_classes = soup.find_all(class_=True)
        
        # Extract and collect all unique class names
        class_names = set()
        for element in elements_with_classes:
            # Each element's class_ attribute is a list of class names
            class_names.update(element['class'])
        
        return class_names
    except requests.exceptions.RequestException as e:
        print(f"Error fetching the webpage: {e}")
        return set()
    except Exception as e:
        print(f"An error occurred: {e}")
        return set()
get_all_classes("https://www.reuters.com/world/europe/death-toll-german-christmas-market-car-ramming-rises-four-bild-reports-2024-12-21/")
```

```python
# We will use the gpt2 tokenizer for the pre-tokenization.
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
# Then we comute the frequencies of each word in the corpus as we do the pre-tokenization:
from collections import defaultdict

word_freqs = defaultdict(int)

for text in result:
    words_with_offsets = AutoTokenizer.from_pretrained("gpt2").backend_tokenizer.pre_tokenizer.pre_tokenize_str(text)
    new_words = [word for word, offset in words_with_offsets]
    for word in new_words:
        word_freqs[word] += 1
```

```python
# Here we compute the base vocabulary, formed by all the characters used in the corpus:
alphabet = []

for word in word_freqs.keys():
    for letter in word:
        if letter not in alphabet:
            alphabet.append(letter)
alphabet.sort()
```

```python
# Add the special tokens used by the model at the beginning of that vocabulary
vocab = ["<|endoftext|>"] + alphabet.copy()
```

```python
# Spliting each word to start training
splits = {word: [c for c in word] for word in word_freqs.keys()}
```

```python
# Computes the frequency of each pair.
# We use this function in each step of the training
def compute_pair_freqs(splits):
    pair_freqs = defaultdict(int)
    for word, freq in word_freqs.items():
        split = splits[word]
        if len(split) == 1:
            continue
        for i in range(len(split) - 1):
            pair = (split[i], split[i + 1])
            pair_freqs[pair] += freq
    return pair_freqs
```

```python
merges = {("ƒ†", "t"): "ƒ†t"}
vocab.append("ƒ†t")
```

```python
def merge_pair(a, b, splits):
    for word in word_freqs:
        split = splits[word]
        if len(split) == 1:
            continue

        i = 0
        while i < len(split) - 1:
            if split[i] == a and split[i + 1] == b:
                split = split[:i] + [a + b] + split[i + 2 :]
            else:
                i += 1
        splits[word] = split
    return splits
```

```python
vocab_size = 100000

while len(vocab) < vocab_size:
    pair_freqs = compute_pair_freqs(splits)
    best_pair = ""
    max_freq = None
    for pair, freq in pair_freqs.items():
        if max_freq is None or max_freq < freq:
            best_pair = pair
            max_freq = freq
    splits = merge_pair(*best_pair, splits)
    merges[best_pair] = best_pair[0] + best_pair[1]
    vocab.append(best_pair[0] + best_pair[1])
```

```python
def tokenize(text):
    pre_tokenize_result = tokenizer._tokenizer.pre_tokenizer.pre_tokenize_str(text)
    pre_tokenized_text = [word for word, offset in pre_tokenize_result]
    splits = [[l for l in word] for word in pre_tokenized_text]
    for pair, merge in merges.items():
        for idx, split in enumerate(splits):
            i = 0
            while i < len(split) - 1:
                if split[i] == pair[0] and split[i + 1] == pair[1]:
                    split = split[:i] + [merge] + split[i + 2 :]
                else:
                    i += 1
            splits[idx] = split

    return sum(splits, [])
```

```python
# Load dataset
Text = pd.read_csv("News_Data.csv")

# Function to preprocess the text
def preprocess_text(text):
    """Cleans the text by removing extra spaces and punctuation, and converts to lowercase."""
    text = re.sub('<[^>]*>', '', str(text))
    emoticons = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)', text)
    text = (re.sub('[\W]+', ' ', text.lower()) +  ' '.join(emoticons).replace('-', ''))
    return text.lower()

# Function to extract patterns from a single article
def extract_patterns_from_text(text, min_words=10):
    """
    Extracts n-word patterns from a given text.
    Args:
        text (str): The input text.
        min_words (int): Minimum number of words per pattern.

    Returns:
        List[str]: Extracted patterns.
    """
    sentences = sent_tokenize(text)
    patterns = []
    for sentence in sentences:
        words = word_tokenize(sentence)
        if len(words) >= min_words:
            patterns.extend([' '.join(words[i:i + min_words]) for i in range(len(words) - min_words + 1)])
    return patterns

# Function to find common patterns across articles
def find_common_patterns(articles, min_words=10):
    """
    Identifies patterns of at least `min_words` that are repeated across multiple articles.
    Args:
        articles (iterable): Collection of article texts.
        min_words (int): Minimum number of words per pattern.

    Returns:
        dict: Dictionary with patterns as keys and their counts as values.
    """
    pattern_counts = defaultdict(int)

    # Use tqdm for a progress bar
    for article in tqdm(articles, desc="Processing Articles"):
        preprocessed_text = preprocess_text(article)
        patterns = extract_patterns_from_text(preprocessed_text, min_words)
        for pattern in patterns:
            pattern_counts[pattern] += 1

    # Filter patterns that occur more than once
    return {pattern: count for pattern, count in pattern_counts.items() if count > 1}

# Function to remove patterns from Text
def remove_patterns_from_text(text, patterns):
    """
    Removes specified patterns from the text.
    Args:
        text (str): The input text.
        patterns (list): List of patterns to remove.

    Returns:
        str: Text with patterns removed.
    """
    for pattern in patterns:
        text = re.sub(re.escape(pattern), "", text)  # Escape pattern for regex safety
    return text

# Main execution
if __name__ == "__main__":
    # min_words_list = [10, 20, 30, 40, 50, 70, 80, 90, 100, 200]  # List of min_words values
    
    Text = pd.read_csv("News_Data.csv")
    
    # Step 1: Find common patterns with a minimum word threshold
    print("Finding common patterns...")
    common_patterns = find_common_patterns(Text["Text"], min_words=30)

    # Step 2: Collect patterns based on occurrence threshold
    Delet_text = []
    print("Collecting patterns with high frequency...")
    for pattern, count in common_patterns.items():
        if count > 4:
            Delet_text.append(pattern)

    # Step 3: Remove patterns from Text["Text"]
    print("Removing patterns from Text...")
    for i in tqdm(range(len(Text["Text"])), desc="Cleaning Text"):
        Text["Text"].iloc[i] = remove_patterns_from_text(Text["Text"].iloc[i], Delet_text)

    # Step 4: Save the updated DataFrame
    Text.to_csv("News_Data.csv", index=False)
    print("Updated News_Data_Updated.csv saved successfully!")
```
